First, let us see what type of dataset we are working with.

```python
df = pd.read_csv('../data/raw/tiak.csv')
df.head()
print(df.dtypes[['rank', 'programme', 'channel', 'start_time', 'end_time', 'rating', 'share', 'date']])
df.info()
```

This initial dataset includes the backbone of our final dataset which includes important features like *rank*, *programme*, *rating*, *share*, etc.

Now, we shall get rid of the programmes that are not affiliated with TV shows in any way, excluding the important sports programmes (as the may influence the ratings of the TV shows). And for the sake of simplification, we will not take summary episodes into account.

```python
size_before_dropping = len(df)

keywords_to_drop = ["TKR","T.S", "OZET"]
rows_to_drop = []

for i in range(len(df)):
  current_title = df.loc[i, 'programme']
  for keyword in keywords_to_drop:
    if keyword.upper() in current_title:
      rows_to_drop.append(i)
      break
df = df.drop(rows_to_drop);

print(f"{size_before_dropping - len(df)} redundant rows deleted.")
```

With the bulk of the redundant data out, at this point, we observe that a show airing on a random hour of a random day, on a random month of a random year does not really mean much in the eyes of a machine. Thus, we will instead create _day_of_week_, _weekday_, and _prime-time (2000-2300 hours)_ features to be utilised.

```python
#TODO
import datetime, time

features_to_add = ["day_of_week", "weekday", "prime-time"]

df[features_to_add] = None

for i in range(len(df)):
  row = df.loc[i]
  dt = datetime.datetime(row.date)
  dt = datetime.strptime(row.time, "%H:%M:%S")

  df.loc[i, "day_of_week"] = dt.weekday()
  df.loc[i, "weekday"] = int(1) if dt.today().weekday() else int(0)
  df.loc[i, "prime_time"] = int(1) if time(20,0) <= dt.time() <= time(2, 0) else int(0)

```
